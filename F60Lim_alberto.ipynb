```python
F60Lim['Incr'] = F60Lim['Eq'].pct_change()
F60Lim.head()
```

```python
F60Lim = F60Lim[F60Lim['Incr'] != 0]
F60Lim.head()
```

```python
# Are you sure you want to do this?
F60Lim['Incr2'] = F60Lim['Incr'].shift(-1)
F60Lim.head()
```

```python
F60Lim['Peak']= (F60Lim['Incr'] * F60Lim['Incr2']) < 0
F60Lim.head()
```

```python
F60Lim.plot(subplots=True, figsize=(12,10), title = 'Equity')
```

```python
F60Lim2 = F60Lim[(F60Lim.Peak == True)]
F60Lim2.head()
```

```python
# Are you sure you want to do this?
F60Lim2['EqAnt'] = F60Lim2['Eq'].shift(1)
F60Lim2.head()
```

```python
F60Lim2['FyH'] = F60Lim2.index
F60Lim2.head()
```

```python
# Are you sure you want to do this?
F60Lim2['FyHAnt'] = F60Lim2['FyH'].shift(1)
F60Lim2.head()
```

```python
F60Lim2['Lapso'] = F60Lim2['FyH'] - F60Lim2['FyHAnt']
F60Lim2.head()
```

```python
F60Lim2['GyP'] = F60Lim2['Eq'] - F60Lim2['EqAnt']
F60Lim2.head()
```

```python
F60Lim2.to_csv('F60Lim2.csv', sep=";")
```
